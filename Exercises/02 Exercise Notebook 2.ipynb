## 2b) Fit an exponential model to the data 

From our knowledge of how the virus spreads, we know that the number of infections, hospital admissions etc. should (roughly) follow an exponential curve. We would therefore like to fit a model of the form $y = c_1 e^{c_2 x}$, where $y$ is the number of admissions on day $x$. 

Note that this is a linear model on a log-scale for $y$. That is, $\log y = \log c_1 +  c_2 x$.

- Fit a linear model for $\log(\text{yvals})$ vs $\text{xvals}$, and print the values of $c_1$ and $c_2$
- Plot the fit $y = c_1 e^{c_2x}$ along with the scatterplot of the data

```python
# your code here

#x now is just x
#y now is log10 y
x = xvals
y = np.log(yvals)

#fit the curve
all_ones = np.ones(np.shape(x))
X = np.column_stack((all_ones,x))
beta = np.linalg.lstsq(X, y, rcond=None)[0] 
print(beta)
c_1 = np.exp(beta[0])
c_2 = beta[1]

plt.scatter(xvals, yvals, s=10)   # s can be used to adjust the size of the dots
plt.plot(xvals, c_1*np.exp(c_2*x), color = 'black', label = 'k=2')
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('Exercise-2-2.pdf', bbox_inches = 'tight')
plt.show()
```

## 2c) Estimate the weekly growth rate in hospital admissions (in \%) over this period

_Hint_ : According to the model, admissions increase every 7 days by a factor of 
$(c_1 e^{c_2 (x+7)})/(c_1 e^{c_2 x}) = e^{7c_2}$.

```python
# compute and print weekly growth rate (in %)
weekly_increase_rate = math.exp(7*c_2) - 1
factor = math.exp(7*c_2)
print("Weekly increase factor is: ", factor*100,"%")
print("Weekly increase rate is: ", weekly_increase_rate*100,"%")
```

## Extension) Use better model to fit the data because, obviously, this is not exponential curve.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def sigmoid(x, L, x0, k, b):
    """
    x  : independent variable
    L  : Sigmoid amplitude
    x0 : middle point
    k  : gradient
    b  : base
    """
    return L / (1 + np.exp(-k * (x - x0))) + b
```

```python
# p0 is initial guess
initial_guess = [max(yvals) - min(yvals),  # initial guess L
                 np.median(xvals),         # initial guess x0 
                 1,                         # initial guess k 
                 min(yvals)]               # initial guess b

popt, pcov = curve_fit(sigmoid, xvals, yvals, p0=initial_guess)

# popt coefficients
# pcov covariance
print("Optimal parameters:", popt)
```

```python
# graph
y_fit = sigmoid(xvals, *popt)

plt.figure(figsize=(8, 5))
plt.scatter(xvals, yvals, label='Data', color='red')
plt.plot(xvals, y_fit, label='Fitted sigmoid', color='blue')
plt.title("Sigmoid Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
